```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

from datetime import *
from datetime import time
from astral import Astral
```

```python
sleep_df = pd.read_excel('../data/cleaned_sleep_data.xlsx', index_col=0)
sleep_df['Start Date'] = pd.to_datetime(sleep_df['Start Date']).dt.date
sleep_df['Start Time'] = pd.to_datetime(sleep_df['Start Time']).dt.time


steps_df = pd.read_excel('../data/cleaned_daily_step_data.xlsx', index_col=0)
steps_df['Date'] = pd.to_datetime(steps_df['Date']).dt.date

weather_df = pd.read_excel('../data/cleaned_weather_data.xlsx', index_col=0)
weather_df['Date'] = pd.to_datetime(weather_df['Date']).dt.date
```

## Merging sleep and steps

```python
merged_df = pd.merge(sleep_df, steps_df, how='inner', left_on='Start Date', right_on='Date')
merged_df.drop(columns='Date', inplace=True)
merged_df
```

## Merging weather

```python
# Weather data merge
display(weather_df)

merged_df = pd.merge(merged_df, weather_df, how='inner', left_on='Start Date', right_on='Date')
display(merged_df)
```

## Adding daylight hours

```python
def get_daylight_hours(row):

    a = Astral()
    a.solar_depression = 'civil'
    city = a['London']
    sun = city.sun(date=row['Date'], local=True)
    
    daylight_amount = sun['sunset'] - sun['sunrise']
    daylight_hours = (daylight_amount.seconds) /3600
    
    return daylight_hours

merged_df['Daylight Hours'] = merged_df.apply(get_daylight_hours, axis=1)
```

## Adding day data

```python
merged_df.insert(loc=2, column='Day', value=pd.to_datetime(merged_df['Date']).dt.dayofweek)
day_df = pd.get_dummies(merged_df.Day, prefix='Day')
merged_df = pd.merge(merged_df, day_df, how='inner', left_index=True, right_index=True)
merged_df
```

```python
columns_to_drop = ['Start Time', 'Day', 'End Date', 'End Time',
                   'Alarm mode', 'Movements per hour', 'Sleep Quality',
                   'Time in bed (seconds)', 'Time asleep (seconds)',
                   'Time before sleep (seconds)', 'Did snore', 'Snore time',
                   'Date']

merged_df = merged_df.drop(columns=columns_to_drop)
merged_df.reset_index(inplace=True, drop=True)
merged_df
```
